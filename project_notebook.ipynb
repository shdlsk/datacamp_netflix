## **Project goals**

Perform exploratory data analysis on the **netflix_data.csv** data to understand more about movies from the 1990s decade.

What was the most frequent movie duration in the 1990s? Save an approximate answer as an integer called **duration** (use 1990 as the decade's start year).

A movie is considered short if it is less than 90 minutes. Count the number of short action movies released in the 1990s and save this integer as **short_movie_count**.



```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
#What was the most frequent movie duration in the 1990s? Save an approximate answer as an integer called duration (use 1990 as the decade's start year).

# Filter data for movies released in the 1990s
movies_1990s = netflix_df[
    (netflix_df['type'] == 'Movie') & 
    (netflix_df['release_year'] >= 1990) & 
    (netflix_df['release_year'] <= 1999)
].copy()

# Convert duration to numeric, extracting numbers (if needed)
movies_1990s['duration'] = pd.to_numeric(movies_1990s['duration'], errors='coerce')

# Find the most frequent duration
most_frequent_duration = int(movies_1990s['duration'].mode()[0])

# Save the result in the original dataframe
duration = most_frequent_duration

print(duration)
```

```python
# Filter data for short action movies released in the 1990s
short_action_movies_1990s = netflix_df[
    (netflix_df['type'] == 'Movie') & 
    (netflix_df['release_year'] >= 1990) & 
    (netflix_df['release_year'] <= 1999) & 
    (netflix_df['duration'] < 90) & 
    (netflix_df['genre'] == 'Action')
].copy()

# Count the number of short action movies
short_movie_count = len(short_action_movies_1990s)
```

```python
# Convert duration to numeric
movies_1990s['duration'] = pd.to_numeric(movies_1990s['duration'], errors='coerce')

# Plot the distribution of durations
plt.figure(figsize=(10, 6))
sns.histplot(movies_1990s['duration'].dropna(), bins=20, kde=True, color='blue')
plt.title('Distribution of Movie Durations in the 1990s')
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')
plt.show()
```
